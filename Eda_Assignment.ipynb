```python
# the whole code is used to  detect the outliers
q1=df.quantile(0.25)
q3=df.quantile(0.75)

IQR=q3-q1

upper_bound=q1-1.5*IQR
lower_bound=q3-1.5*IQR

outlier_iqr= df[( df < lower_bound) | (df > upper_bound)]

outlier_iqr
```
