```python
x=data.drop('type',axis=1)
y=data['type']
```

```python
# splitting dataset into train and test
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# implementing K-Nearest algorithom
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=5
knn=KNeighborsClassifier(n_neighbors=k,metric='euclidean')

knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Predictions
y_pred = knn.predict(x_test)

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print(classification_report(y_test, y_pred))
```

```python
# visualising decision boundaries
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a mesh grid for the first two features
x_min, x_max = x_train.iloc[:, 0].min() - 1, x_train.iloc[:, 0].max() + 1
y_min, y_max = x_train.iloc[:, 1].min() - 1, x_train.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Create a base array for the 17 features
base_array = np.zeros((xx.ravel().shape[0], 17))
# Set the first two features to the grid values
base_array[:, 0] = xx.ravel()
base_array[:, 1] = yy.ravel()
# Set the remaining features to the mean value of those features in the training set
for i in range(2, 17):
    base_array[:, i] = x_train.iloc[:, i].mean()

# Predict for each point in the grid
Z = knn.predict(base_array)
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
plt.scatter(x_train.iloc[:, 0], x_train.iloc[:, 1], c=y_train, edgecolor='k', marker='o', cmap=plt.cm.coolwarm)
plt.title("Decision Boundaries")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
