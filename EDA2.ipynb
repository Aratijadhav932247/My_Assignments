# data exploration and preprocessing

```python
df.dtypes
```

```python
data_encoded=pd.get_dummies(data,drop_first=True)

for col in data_encoded.select_dtypes(include=['int64', 'object']):
    data_encoded[col] = (data_encoded[col] >= data_encoded[col].median()).astype(int)
    
    
```
